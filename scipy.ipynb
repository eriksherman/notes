### Optimization

Minimize $f(x) = (x-3)^2$

```python
def f(x):
	return (x-3)**2

res = minimize(f,2)
res.x
```

Minimize $f(x,y) = (x-1)^2 + (y-2.5)^2$ subject to...

$x-2y+2 \geq  0$

$-x-2y+6 \geq  0$

$-x+2y+2 \geq  0$

$x \geq 0$

$y \geq 0$



```python
f = lambda x: (x[0] - 1)**2 + (x[1] - 2.5)**2
```

```python
# define constraints
# must be specified as <expression> >= 0
constr = ({'type': 'ineq', 'fun': lambda x: x[0] - 2*x[1]+2 },
{'type': 'ineq', 'fun': lambda x: x[0] - 2*x[1]+2 },
{'type': 'ineq', 'fun': lambda x: -x[0] - 2*x[1]+6 },
{'type': 'ineq', 'fun': lambda x: -x[0] - 2*x[1]+2 })

# x goes from 0 to inf
# y goes from 0 to inf
bounds = ((0, None), (0,None))

res = minimize(f, [2,0], bounds=bounds, constraints=constr)
res.x
```
